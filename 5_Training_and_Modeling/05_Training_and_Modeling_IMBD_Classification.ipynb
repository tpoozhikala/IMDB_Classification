## 5.5 Deep Learning Image Classification Predictive Model

### 5.5.1 Intialize/Train Predictive Model

Prior to training the image classifier model, the metric f1_score is chosen as the performance metric for hyperparameter tuning within GridSearchCV to get an optimal image classifier model performer. F1_score chosen as the metric as it provides a harmonic mean for the model's performance regarding recall and precision regarding predictions for the the multi-genre labels.

```python
#Custom f1_metric scorer created after providing prompt within chatGPT for creating custom f1_metric scorer with Keras backend
def f1_metric_2(y_true, y_pred):
    # Calculate true positives, false positives, and false negatives
    tp = K.sum(y_true * y_pred)
    fp = K.sum((1 - y_true) * y_pred)
    fn = K.sum(y_true * (1 - y_pred))

    # Calculate precision, recall, and F1 score
    precision = tp / (tp + fp + K.epsilon())
    recall = tp / (tp + fn + K.epsilon())
    f1 = 2 * precision * recall / (precision + recall + K.epsilon())
    
    return f1
```

Custom Keras Classifier created due to compatilbilty issues regarding current version of tensorflow and keras where scikeras.wrapper is no longer recognized.

```python
#Code regarding buidling a custom Keras classifier provided by chatGPT when given prompt regarding building custom keras image classifier model
class CustomKerasClassifier_3(BaseEstimator, ClassifierMixin):
    def __init__(self, activation='relu', optimizer='adam', learning_rate=0.001, epochs=10, batch_size=10):
        self.activation = activation
        self.optimizer = optimizer
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        self.model = None
        self.classes_ = None
        
    def fit(self, X, y):
        # Check that X and y have correct shape
        #X, y = check_X_y(X, y)

        # Select optimizer based on the input parameter
        if self.optimizer.lower() == 'sgd':
            opt = SGD(learning_rate=self.learning_rate)
        elif self.optimizer.lower() == 'adagrad':
            opt = Adagrad(learning_rate=self.learning_rate)
        elif self.optimizer.lower() == 'adam':
            opt = Adam(learning_rate=self.learning_rate)
        else:
            raise ValueError("Not Supported Optimizer")
        
        # Define the Keras model
        model = Sequential([
            Conv2D(50, kernel_size=3, activation=self.activation, input_shape=(200, 200, 3)),
            MaxPooling2D(2),
            Dropout(0.20),
            Conv2D(20, kernel_size=3, activation=self.activation),
            MaxPooling2D(2),
            Dropout(0.20),
            Conv2D(10, kernel_size=3, activation=self.activation),
            MaxPooling2D(2),
            Dropout(0.20),
            Flatten(),
            Dense(64, activation=self.activation),
            Dropout(0.5),
            Dense(32, activation=self.activation),
            Dropout(0.5),
            Dense(y.shape[1], activation='sigmoid')
        ])

        # Compile the model with F1 score as the metric
        model.compile(optimizer=opt, loss='binary_crossentropy', metrics=[f1_metric_2])

        # Define callbacks
        callbacks = [EarlyStopping(patience=5)]

        # Fit the model
        model.fit(X, y, epochs=self.epochs, batch_size=self.batch_size, callbacks=callbacks, validation_split=0.2)
        
        # Save the model
        self.model = model

        return self
    
    def predict(self, X):
        # Check is fit had been called
        check_is_fitted(self)
        
        # Input validation
        #X = check_array(X)
        
        # Predict probabilities
        y_pred_proba = self.model.predict(X)
        
        # Convert probabilities to binary predictions
        y_pred = (y_pred_proba > 0.5).astype(int)
        
        return y_pred
    
    def score(self, X, y):
        # Predict labels
        y_pred = self.predict(X)
        
        # Calculate F1 score for each label individually
        f1_scores = []
        for i in range(y.shape[1]):
            f1_scores.append(f1_score(y[:, i], y_pred[:, i]))
        
        # Return the average F1 score across all labels
        return np.mean(f1_scores)
    
    def get_params(self, deep=True):
        return {
            'activation': self.activation,
            'optimizer': self.optimizer,
            'learning_rate': self.learning_rate,
            'epochs': self.epochs,
            'batch_size': self.batch_size
        }
    
    def set_params(self, **params):
        if 'activation' in params:
            self.activation = params['activation']
        if 'optimizer' in params:
            self.optimizer = params['optimizer']
        if 'learning_rate' in params:
            self.learning_rate = params['learning_rate']
        if 'epochs' in params:
            self.epochs = params['epochs']
        if 'batch_size' in params:
            self.batch_size = params['batch_size']
        return self
```

```python
#Code regarding taking in hyperparameters and then fitting the associated model found on chatGPT after given prompt for designing custom image classifier model
from sklearn.model_selection import GridSearchCV

# Define the parameter grid with the specified activation functions and optimizers
param_grid = {
    'activation': ['relu', 'tanh', 'linear'],
    'optimizer': ['SGD', 'Adagrad', 'Adam'],
    'learning_rate': [0.001],
    'epochs': [10],
    'batch_size': [10]
}

# Create an instance of CustomKerasClassifier
keras_classifier = CustomKerasClassifier_3()

# Create GridSearchCV object
grid_search = GridSearchCV(estimator=keras_classifier, param_grid=param_grid, scoring='f1_micro', cv=3)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Print the best parameters found
print("Best parameters found:", grid_search.best_params_)

# Get the best estimator
best_classifier = grid_search.best_estimator_

# Evaluate the best estimator
#score = best_classifier.score(X_test, y_test)
#print("F1 Score on test set:", score)
```

## 5.6 Summary/Future Work

As a summary of the above steps, an image classifier model was developed through the Keras module where several hyperparameters were initialized prior to training or "fit" from the training set. In order to initialize the model, first the model had to take in the proper RGB value array of (200,200,3) meaning 200x200 pixel values for the image and its respective RGB values. Specifically, the first few layers are convolution layers meaning pixel values are scanned with a 3x3 window to receive an output. Then all these values are then followed with a Max Pooling layer to receive the max output to be used in the subsequent layers. Then also there are dropout layers in order to promote a more generalizable model within the training. Then a GridSearchCV was performed in order to hyperparameter tune regarding the image classifier model optimizers and activation functions with a 3-fold cross-validation with a best score metric of f1_score as this metric gives an understanding of overall model performance for predicting the different genres for movies. However, based on the overall f1_score of around 0.25, it seems the model needs improvement before it can be safely deployed regarding providing predictions of genres of movies based on the poster image data. The underperformance as previously mentioned in the above section might be due to several factors including the imbalance of the different genres within the dataset, the number of layers within the image classifier, having a 3-fold versus potentially applying a 5-fold cross validation, and the limited number of hyperparameters checked within the GridSearchCV.

Future work will involve increasing the samples of different genres of movies and corresponding poster RGB image values, adjusting the layers within the image classifier to be more optimal for RGB value input to predict output, and increasing the number of hyperparameters to include different learning rates, optimization functions, and activation functions for the different layers.
